```python
import pandas as pd
from datetime import datetime

# Create a findings dataframe
findings_data = []

# Template for each finding:
# Category, Insight, Evidence, Business Impact, Recommendation

# Save empty template for now - we'll fill it step by step
findings_template = pd.DataFrame(columns=[
    'Category', 
    'Insight', 
    'Evidence', 
    'Business_Impact', 
    'Recommendation'
])

# Save template
findings_template.to_csv('D:/Netflix content strategy analysis/outputs/reports/findings_template.csv', index=False)
print("✅ Findings template created at: outputs/reports/findings_template.csv")
print("\nYou'll fill this with your insights in the next steps.")
```

```python
# Calculate content mix metrics
total_titles = len(df)
movies = len(df[df['type'] == 'Movie'])
tv_shows = len(df[df['type'] == 'TV Show'])
movies_pct = (movies/total_titles)*100
tv_pct = (tv_shows/total_titles)*100

# Create insights list
content_insights = []

# Insight 1: Content Distribution
content_insights.append({
    'Category': 'Content Mix',
    'Insight': f'Netflix library consists of {movies_pct:.1f}% Movies and {tv_pct:.1f}% TV Shows',
    'Evidence': f'Movies: {movies:,}, TV Shows: {tv_shows:,} out of {total_titles:,} total titles',
    'Business_Impact': 'Shows Netflix prioritizes movies over TV series in their library',
    'Recommendation': 'Monitor if this ratio aligns with viewer engagement - consider increasing TV shows if they drive more engagement per title'
})

# Convert to DataFrame
content_df = pd.DataFrame(content_insights)

# Display
print("📊 CONTENT MIX INSIGHTS:")
print("-" * 50)
print(content_df.to_string(index=False))

# Save
content_df.to_csv('D:/Netflix content strategy analysis/outputs/reports/content_mix_insights.csv', index=False)
print("\n✅ Content mix insights saved!")
```

```python
# Process genres for analysis
df_genres = df.copy()
df_genres['genres'] = df_genres['listed_in'].str.split(', ')
genre_exploded = df_genres.explode('genres')
genre_exploded['genres'] = genre_exploded['genres'].str.strip()

# Calculate genre metrics
top_genres = genre_exploded['genres'].value_counts()
unique_genres = genre_exploded['genres'].nunique()

# Create genre insights
genre_insights = []

# Insight 1: Top Genres
top_3 = top_genres.head(3)
genre_insights.append({
    'Category': 'Genre Dominance',
    'Insight': f'Top 3 genres: {top_3.index[0]}, {top_3.index[1]}, and {top_3.index[2]}',
    'Evidence': f'{top_3.index[0]}: {top_3.values[0]} titles, {top_3.index[1]}: {top_3.values[1]} titles, {top_3.index[2]}: {top_3.values[2]} titles',
    'Business_Impact': 'These genres define Netflix\'s core content strategy and brand identity',
    'Recommendation': 'Ensure strong representation in these genres while avoiding oversaturation'
})

# Insight 2: Genre Diversity
genre_insights.append({
    'Category': 'Genre Diversity',
    'Insight': f'Netflix features {unique_genres} unique genre categories',
    'Evidence': f'From {genre_exploded["genres"].nunique()} distinct genre labels',
    'Business_Impact': 'Extensive categorization helps with content discovery and targeting niche audiences',
    'Recommendation': 'Use this granular categorization for personalized recommendations'
})

# Convert to DataFrame
genre_df = pd.DataFrame(genre_insights)

# Display
print("\n🎭 GENRE INSIGHTS:")
print("-" * 50)
print(genre_df.to_string(index=False))

# Save
genre_df.to_csv('D:/Netflix content strategy analysis/outputs/reports/genre_insights.csv', index=False)
print("\n✅ Genre insights saved!")
```

```python
# Determine which country column to use
country_col = 'primary_country' if 'primary_country' in df.columns else 'country'

# Calculate country metrics
top_countries = df[country_col].value_counts()
unique_countries = df[country_col].nunique()

# US vs International
us_titles = df[df[country_col].str.contains('United States', na=False)].shape[0]
us_share = (us_titles / len(df)) * 100
intl_share = 100 - us_share

# Create country insights
country_insights = []

# Insight 1: US Dominance
country_insights.append({
    'Category': 'Geographic Distribution',
    'Insight': f'United States dominates with {us_share:.1f}% of content',
    'Evidence': f'{us_titles:,} out of {len(df):,} titles are from US',
    'Business_Impact': 'Strong home market advantage but potential over-reliance on US content',
    'Recommendation': 'Continue international expansion to reduce dependency on US market'
})

# Insight 2: Top International Markets
top_intl = top_countries[~top_countries.index.str.contains('United States', na=False)].head(3)
if len(top_intl) >= 3:
    country_insights.append({
        'Category': 'International Growth',
        'Insight': f'Top international markets: {top_intl.index[0]}, {top_intl.index[1]}, {top_intl.index[2]}',
        'Evidence': f'{top_intl.index[0]}: {top_intl.values[0]} titles, {top_intl.index[1]}: {top_intl.values[1]} titles, {top_intl.index[2]}: {top_intl.values[2]} titles',
        'Business_Impact': 'These are key growth markets for international expansion',
        'Recommendation': 'Increase investment in local content for these markets'
    })

# Insight 3: Global Reach
country_insights.append({
    'Category': 'Global Presence',
    'Insight': f'Content represents {unique_countries} countries worldwide',
    'Evidence': f'Titles from {unique_countries} different nations',
    'Business_Impact': 'Shows Netflix\'s commitment to global content strategy',
    'Recommendation': 'Continue diversifying geographic representation'
})

# Convert to DataFrame
country_df = pd.DataFrame(country_insights)

# Display
print("\n🌍 COUNTRY INSIGHTS:")
print("-" * 50)
print(country_df.to_string(index=False))

# Save
country_df.to_csv('D:/Netflix content strategy analysis/outputs/reports/country_insights.csv', index=False)
print("\n✅ Country insights saved!")
```

```python
# Calculate temporal metrics
yearly_counts = df['year_added'].value_counts().sort_index()
earliest_year = int(df['year_added'].min())
latest_year = int(df['year_added'].max())
total_years = latest_year - earliest_year + 1

# Calculate growth
pre_2016 = df[df['year_added'] < 2016].shape[0]
post_2016 = df[df['year_added'] >= 2016].shape[0]
growth_pct = ((post_2016 - pre_2016) / pre_2016) * 100

# Peak year
peak_year = yearly_counts.idxmax()
peak_count = yearly_counts.max()

# Create temporal insights
temporal_insights = []

# Insight 1: Growth Period
temporal_insights.append({
    'Category': 'Growth Timeline',
    'Insight': f'Netflix added content over {total_years} years ({earliest_year}-{latest_year})',
    'Evidence': f'Content additions span from {earliest_year} to {latest_year}',
    'Business_Impact': 'Long-term consistent investment in content library',
    'Recommendation': 'Maintain consistent content addition strategy'
})

# Insight 2: Accelerated Growth
temporal_insights.append({
    'Category': 'Growth Rate',
    'Insight': f'Content additions grew by {growth_pct:.0f}% after 2016',
    'Evidence': f'Pre-2016: {pre_2016} titles, Post-2016: {post_2016} titles',
    'Business_Impact': 'Major strategic shift towards aggressive content acquisition',
    'Recommendation': 'Focus on quality control to match increased quantity'
})

# Insight 3: Peak Year
temporal_insights.append({
    'Category': 'Peak Production',
    'Insight': f'{int(peak_year)} was the peak year with {peak_count} titles added',
    'Evidence': f'Highest single-year content addition in Netflix history',
    'Business_Impact': 'Represents maximum content investment period',
    'Recommendation': 'Analyze viewer engagement from peak years to inform future strategy'
})

# Convert to DataFrame
temporal_df = pd.DataFrame(temporal_insights)

# Display
print("\n📅 TEMPORAL INSIGHTS:")
print("-" * 50)
print(temporal_df.to_string(index=False))

# Save
temporal_df.to_csv('D:/Netflix content strategy analysis/outputs/reports/temporal_insights.csv', index=False)
print("\n✅ Temporal insights saved!")
```
