```python
# Cell 1: Import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Display settings for better viewing
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("✅ Libraries imported successfully")
```

```python
# Cell 2: Load the Netflix dataset
# Make sure your CSV file is in the 'data' folder
df = pd.read_csv('D:/Netflix content strategy analysis/data/raw data/netflix_titles.csv')

# Quick check
print(f"✅ Dataset loaded successfully!")
print(f"📊 Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\n📋 First 3 rows:")
df.head(3)
```

```python
# Cell 3: Check basic info about your data
print("📋 DATA INFORMATION:")
print("-" * 50)
df.info()
```

```python
# Cell 4: Check for missing values
print("🔍 MISSING VALUES COUNT:")
print("-" * 50)
missing_data = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Create a nice dataframe to view missing data
missing_df = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': missing_data.values,
    'Missing_Percentage': missing_percentage.values.round(2)
})
print(missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False))
```

```python
# Cell 5: Check basic statistics of numerical columns
print("📊 BASIC STATISTICS:")
print("-" * 50)
# First, let's see if we have any numeric columns to check
df.describe()
```

```python
# Cell 6: Check unique values in key columns
print("🎯 UNIQUE VALUES IN KEY COLUMNS:")
print("-" * 50)

categorical_columns = ['type', 'rating', 'country', 'listed_in']
for col in categorical_columns:
    unique_count = df[col].nunique()
    print(f"{col}: {unique_count} unique values")
    
    # Show top 5 most common values for each
    print(f"Top 5 in {col}:")
    print(df[col].value_counts().head(3))
    print("-" * 30)
```

```python
# Cell 7: Create a clean copy of the dataframe
df_clean = df.copy()
print("✅ Created clean copy of dataframe")
```

```python
# Cell 8: Handle missing values in 'director' column
print("🎬 Handling missing directors...")
# Check current missing count
print(f"Before: {df_clean['director'].isnull().sum()} missing")

# Fill with 'Unknown'
df_clean['director'] = df_clean['director'].fillna('Unknown')

# Verify
print(f"After: {df_clean['director'].isnull().sum()} missing")
print("✅ Director column cleaned")
```

```python
# Cell 9: Handle missing values in 'cast' column
print("👥 Handling missing cast members...")
print(f"Before: {df_clean['cast'].isnull().sum()} missing")

# Fill with 'Unknown'
df_clean['cast'] = df_clean['cast'].fillna('Unknown')

print(f"After: {df_clean['cast'].isnull().sum()} missing")
print("✅ Cast column cleaned")
```

```python
# Cell 10: Handle missing values in 'country' column
print("🌍 Handling missing countries...")
print(f"Before: {df_clean['country'].isnull().sum()} missing")

# Let's find the most common country (mode)
most_common_country = df_clean['country'].mode()[0]
print(f"Most common country: {most_common_country}")

# Fill missing with the most common country
df_clean['country'] = df_clean['country'].fillna(most_common_country)

print(f"After: {df_clean['country'].isnull().sum()} missing")
print("✅ Country column cleaned")
```

```python
# Cell 11: Handle 'date_added' and 'rating' - these are important for analysis
print("📅 Handling date_added and rating...")

# Check current state
print(f"date_added missing: {df_clean['date_added'].isnull().sum()}")
print(f"rating missing: {df_clean['rating'].isnull().sum()}")

# Remove rows where either is missing (they're crucial for analysis)
initial_rows = len(df_clean)
df_clean = df_clean.dropna(subset=['date_added', 'rating'])
final_rows = len(df_clean)

print(f"Rows removed: {initial_rows - final_rows}")
print(f"Rows remaining: {final_rows}")
print("✅ Cleaned date_added and rating")
```

```python
# Cell 12: Clean and convert date_added
print("🔄 Converting date_added to datetime format...")

# Check a sample of date_added before cleaning
print("Sample of raw date_added values:")
print(df_clean['date_added'].head())

# Remove any extra spaces
df_clean['date_added'] = df_clean['date_added'].str.strip()

# Convert to datetime
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'])

# Check the result
print("\nAfter conversion:")
print(df_clean['date_added'].head())
print(f"\nDate range: {df_clean['date_added'].min()} to {df_clean['date_added'].max()}")
print("✅ Date column converted successfully")
```

```python
# Cell 13: Extract temporal features from date_added
print("📊 Creating temporal features...")

# Extract year, month, quarter
df_clean['year_added'] = df_clean['date_added'].dt.year
df_clean['month_added'] = df_clean['date_added'].dt.month
df_clean['quarter_added'] = df_clean['date_added'].dt.quarter

# Extract day name (Monday, Tuesday, etc.)
df_clean['day_added'] = df_clean['date_added'].dt.day_name()

# Check the new columns
print("New columns created:")
print(df_clean[['date_added', 'year_added', 'month_added', 'quarter_added', 'day_added']].head())
print("✅ Temporal features created")
```

```python
# Cell 14: Create decade feature from release_year
print("📅 Creating release decade feature...")

# Create decade (e.g., 1990, 2000, 2010)
df_clean['release_decade'] = (df_clean['release_year'] // 10) * 10

# Check distribution
print("\nContent by decade:")
print(df_clean['release_decade'].value_counts().sort_index())
print("✅ Decade feature created")
```

```python
# Cell 15: Calculate content age at addition
print("⏱️ Calculating content age at addition...")

# Calculate how old content was when added to Netflix
df_clean['content_age_at_addition'] = df_clean['year_added'] - df_clean['release_year']

# Check some examples
print("Examples of content age:")
example = df_clean[['title', 'release_year', 'year_added', 'content_age_at_addition']].head(10)
print(example)

# Basic statistics
print(f"\nAverage age at addition: {df_clean['content_age_at_addition'].mean():.1f} years")
print(f"Oldest content added: {df_clean['content_age_at_addition'].max()} years old")
print("✅ Content age feature created")
```

```python
# Cell 16: Process 'listed_in' (genres) column
print("🎭 Processing genres column...")

# Count how many genres each title has
df_clean['genre_count'] = df_clean['listed_in'].apply(lambda x: len(str(x).split(',')))

# Show results
print("Genre count distribution:")
print(df_clean['genre_count'].value_counts().sort_index())

# Example of titles with multiple genres
print("\nExample of titles with most genres:")
multi_genre = df_clean.nlargest(5, 'genre_count')[['title', 'listed_in', 'genre_count']]
print(multi_genre)
print("✅ Genre count feature created")  
```

```python
# Cell 17: Process 'cast' column to count cast members
print("👥 Counting cast members...")

# Function to count cast members safely
def count_cast_members(cast_string):
    if pd.isna(cast_string) or cast_string == 'Unknown':
        return 0
    return len(str(cast_string).split(','))

# Apply the function
df_clean['cast_count'] = df_clean['cast'].apply(count_cast_members)

# Check distribution
print("Cast size distribution:")
print(df_clean['cast_count'].value_counts().sort_index().head(10))

print(f"\nAverage cast size: {df_clean['cast_count'].mean():.1f} members")
print(f"Maximum cast size: {df_clean['cast_count'].max()} members")
print("✅ Cast count feature created")
```

```python
# Cell 18: Process duration for Movies
print("🎬 Processing movie durations...")

# Create a column for movie durations in minutes
def extract_movie_duration(duration_string):
    if pd.isna(duration_string):
        return None
    if 'min' in str(duration_string):
        return int(str(duration_string).replace(' min', ''))
    return None

df_clean['duration_minutes'] = df_clean['duration'].apply(extract_movie_duration)

# Check movie durations
movie_durations = df_clean[df_clean['type'] == 'Movie']['duration_minutes'].dropna()
print(f"Movies: {len(movie_durations)} titles with duration info")
print(f"Average movie length: {movie_durations.mean():.0f} minutes")
print(f"Shortest movie: {movie_durations.min()} minutes")
print(f"Longest movie: {movie_durations.max()} minutes")
print("✅ Movie duration feature created")
```

```python
# Cell 19: Process duration for TV Shows
print("📺 Processing TV show seasons...")

# Create a column for TV show seasons
def extract_tv_seasons(duration_string):
    if pd.isna(duration_string):
        return None
    if 'Season' in str(duration_string):
        # Handle "1 Season" or "2 Seasons"
        return int(str(duration_string).split()[0])
    return None

df_clean['seasons'] = df_clean['duration'].apply(extract_tv_seasons)

# Check TV show seasons
tv_seasons = df_clean[df_clean['type'] == 'TV Show']['seasons'].dropna()
print(f"TV Shows: {len(tv_seasons)} titles with season info")
print(f"Average seasons: {tv_seasons.mean():.1f}")
print(f"Most seasons: {tv_seasons.max()}")
print("✅ TV show seasons feature created")
```

```python
# Cell 20: Extract primary country (first country listed)
print("🌍 Creating primary country feature...")

def extract_primary_country(country_string):
    if pd.isna(country_string) or country_string == 'Unknown':
        return 'Unknown'
    # Take first country if multiple are listed
    return str(country_string).split(',')[0].strip()

df_clean['primary_country'] = df_clean['country'].apply(extract_primary_country)

# Check top primary countries
print("Top 10 primary countries:")
print(df_clean['primary_country'].value_counts().head(10))
print("✅ Primary country feature created")
```

```python
# Cell 21: Final check of cleaned data
print("🔍 FINAL DATA QUALITY CHECK")
print("-" * 50)

# Check for any remaining missing values
print("Remaining missing values:")
remaining_missing = df_clean.isnull().sum()
print(remaining_missing[remaining_missing > 0])

# Check new columns created
print("\nNew columns added:")
original_columns = set(df.columns)
new_columns = set(df_clean.columns) - original_columns
for col in sorted(new_columns):
    print(f"  ✅ {col}")

print(f"\nFinal shape: {df_clean.shape[0]} rows × {df_clean.shape[1]} columns")
```

```python
# Cell 22: Save the cleaned dataset
print("💾 Saving cleaned data...")

# Save to CSV
df_clean.to_csv('D:/Netflix content strategy analysis/data/processed data/netflix_clean_data.csv', index=False)
print("✅ Cleaned data saved to 'data/netflix_clean_data.csv'")

# Also save a sample for quick viewing
df_clean.head(10).to_csv('D:/Netflix content strategy analysis/data/sample data/sample_cleaned_data.csv', index=False)
print("✅ Sample data saved to 'data/sample_cleaned_data.csv'")
```
